```python
import numpy as np
import pandas as pd
import requests
import json

url = "https://data.covid19india.org/v4/min/timeseries.min.json"
JSONContent = requests.get(url).json()
```

```python
states = []
dates = []
dic = {}

json = JSONContent

for state in json:
    states.append(state)
    x = json[state]
    for i in x['dates']:
        dates.append(i)
#print(dates)
#print(states)
for currdate in dates:
    dic[currdate] = {}
    
for currstate in  states:
    for currdate in dates:
        if currdate in json[currstate]['dates']:
            
            val = json[currstate]['dates'][currdate]
            if 'delta' in val:
                x = val['delta']
                if 'confirmed' in x:
                    dic[currdate][currstate]  = x['confirmed']
                    #print(x['confirmed'])
                else:
                    dic[currdate][currstate]  = 0
                    #print(0)
            else:
                dic[currdate][currstate]  = 0
                #print(0)
        else:
            dic[currdate][currstate]  = 0
             #print(0)
#print(dic)
```

```python
nd = {}
f = 0
for state in states:
    nd[state] = []
    for currdate in dates:
        nd[state].append(dic[currdate][state])
        f+=1
#print(f)        
#print(nd['WB'])



        
```

```python
ndframe = pd.DataFrame.from_dict(nd)
ndframe['date'] = dates
```
